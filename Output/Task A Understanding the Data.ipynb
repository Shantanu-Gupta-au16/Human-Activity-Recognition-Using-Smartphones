```python
#Task A
#Understanding the Data
1-The objective of this paper is how to recognize certain types of human physical activities like walking,sitting,walking_upstairs,
walking_downstairs,standing,laying using mobile sensor's data of Accelerometer and Gyroscope generated by a user's cell phone.Used 
some pre-processing steps on the data collected by mobile sensor's and applied Machine learning algorithm,for predicting whether the
human is walking,going upstairs or downstairs,laying down or just sitting.I think this concept is somewhat related to real fitness bands
devices like Fitbit that measures overall human activity.
```

```python
2-Human Activity Type:
1 WALKING
2 WALKING_UPSTAIRS
3 WALKING_DOWNSTAIRS
4 SITTING
5 STANDING
6 LAYING

A group of 30 volunteers with ages ranging from 19 to 48 years were selected for this task.
```

```python
3-X_train:7352 rows and 561 features(columns)
X_test:2947 rows and 561 features(columns)
y_train:7352 rows and 1 features that is Label
y_test:2947 rows and 1 features that is Label

A total of 561 features were extracted to describe each activity.    
```

```python
4-Multiclass Support Vector Machine (SVM) - One-Vs-All (OVA) approach is used.The obtained dataset has been randomly partitioned into two sets, where 70% of the volunteers was selected for generating the training data and 30% the test data.
Overall accuracy of 96% for the test data.
```
